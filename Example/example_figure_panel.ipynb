Style Sheet
This cell defines the shared plotting style used throughout the notebook.

It should be placed near the top of the script before creating any figures, so that all plots use the same:
- font family and sizes
- line and marker styling
- axis and tick appearance
- figure resolution
- PDF font settings

It also includes a helper function for adding panel labels to subplots.

```python
from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib import image as mpimg
import numpy as np

# Define a consistent style for all figures
plt.rcParams.update({
    "font.family": "sans-serif",
    # "font.sans-serif": ['DejaVu Sans', 'Bitstream Vera Sans', 'Computer Modern Sans Serif', 'Lucida Grande', 'Verdana', 'Geneva', 'Lucid', 'Arial', 'Helvetica', 'Avant Garde', 'sans-serif'],
    "font.sans-serif": ['DejaVu Sans','TeX Gyre Heros'],
    "font.size": 6,
    "figure.labelsize": 7,
    "axes.labelsize": 7,
    "axes.titlesize": 7,
    "xtick.labelsize": 6,
    "ytick.labelsize": 6,
    "axes.linewidth": 0.6,
    "lines.linewidth": 0.6,
    "lines.markersize": 2.0,
    "lines.markeredgewidth": 0.5,
    "errorbar.capsize": 2,
    "patch.linewidth": 0.6,
    "xtick.major.width": 0.6,
    "ytick.major.width": 0.6,
    "xtick.major.size": 2,
    "ytick.major.size": 2,
    "legend.fontsize": 6,
    "figure.dpi": 300,
    "pdf.fonttype": 42,
})

def letter_annotation(ax, letter):
    ax.annotate(
        letter,
        xy=(0, 1),
        xycoords='axes fraction',
        xytext=(-20, 6),
        textcoords='offset points',
        ha='right',
        va='bottom',
        fontsize=7,
        fontweight='bold',
        clip_on=False,
    )
```

```python
def drawfigureB(fig):
    # ========================== Loading data from CSV files ===========================================
    x1, y1 = np.loadtxt('./FigureB/AEA.csv', delimiter=',', skiprows=1, unpack=True)
    x2, y2 = np.loadtxt('./FigureB/BA.csv', delimiter=',', skiprows=1, unpack=True)
    x3, y3 = np.loadtxt('./FigureB/PEA.csv', delimiter=',', skiprows=1, unpack=True)
    x4, y4 = np.loadtxt('./FigureB/SEA.csv', delimiter=',', skiprows=1, unpack=True)

    ax = fig.subplots(1,1)
    ax.plot(x1, y1, 'o', label='Data', markerfacecolor='none',color='C0')
    ax.plot(x1,y1, '-', label='Fit',color='C0')
    ax.plot(x2, y2, 'o', label='Data', markerfacecolor='none',color='C1')
    ax.plot(x2,y2, '-', label='Fit',color='C1')
    ax.plot(x3, y3, 'o', label='Data', markerfacecolor='none',color='C2')
    ax.plot(x3,y3, '-', label='Fit',color='C2')
    ax.plot(x4, y4, 'o', label='Data', markerfacecolor='none',color='C3')
    ax.plot(x4,y4, '-', label='Fit',color='C3')
    # ax.plot(x1, y_fit1, '-', label='Fit',color='C1')
    # ax.plot(x2, y_fit2, '-', label='Fit 2',color='C2')
    # ax.set_yscale('log')
    ax.set_xscale('log')
    ax.set_xlabel('Delay (ps)')
    ax.set_ylabel('Ratio')
    ax.tick_params(direction='in', top=True, right=True,which='both')
    # ax.legend(frameon=False, loc='best')
    ax.text(1700, 0.45, 'AEA', ha='center', va='center',color='C0')
    ax.text(1700, 0.90, 'BA', ha='center', va='center',color='C1')
    ax.text(1700, 0.6, 'PEA', ha='center', va='center',color='C2')
    ax.text(1700, 0.70, 'SEA', ha='center', va='center',color='C3')
    letter_annotation(ax, 'B')

drawfigureB(plt.figure(figsize=(4,3)))
```

```python
def drawfigureE(fig):
    
    #------------------------------------ Loading data from CSV files --------------------------------------
    x_left, y_left, y_left_err = np.loadtxt('./FigureE/red_lambda_curve.csv', delimiter=',', skiprows=1, unpack=True)
    x_right, y_right, y_right_err = np.loadtxt('./FigureE/blue_g_parameter_curve.csv', delimiter=',', skiprows=1, unpack=True)
    
    axL = fig.subplots(1,1)
    
    #------------------------------- Left Y-axis -----------------------------------------------------------------------
    color = 'C0'
    
    axL.errorbar(x_left, y_left, yerr=y_left_err, color=color, fmt='o', markerfacecolor='none', label='$\\Lambda$')
    # axL.set_xlim(left=-5, right=60)
    axL.set_ylim(bottom=0.03,top=0.12)
    axL.set_xlabel("Thickness (nm)")
    axL.set_ylabel("$\\Lambda$ (Wm$^{-1}$K$^{-1}$)", color=color)
    axL.text(120, 0.11, 'AEA, $T_{a}$ = 80 $^{\\circ}$C', color='black', ha='center', va='center',rotation=0, transform_rotates_text=True)
    
    # Graphic adjustments for left Y-axis (don't edit if you don't know what it does)
    axL.spines['left'].set_color(color)
    axL.tick_params(direction='in', top=True, right=True)
    axL.tick_params(axis='y',colors=color)

    #--------------------------------- Right Y-axis ------------------------------------------------------------------
    axR = axL.twinx()
    color = 'C1'
    
    axR.errorbar(x_right, y_right, yerr=y_right_err, color=color, fmt='^', markerfacecolor='none', label='g parameter')
    axR.set_ylim(bottom=0,top=10)
    axR.set_ylabel("g parameter (%)", color=color)
    # axR.text(40, 400, 'Cumulative \n Heat Capacity', color=color, ha='center', va='center',rotation=0, transform_rotates_text=True)
    
    
    # Graphic adjustments for right Y-axis (don't edit if you don't know what it does)
    axR.spines['left'].set_visible(False)
    axR.spines['right'].set_color(color)
    axR.spines['top'].set_visible(False)
    axR.spines['bottom'].set_visible(False)
    axR.tick_params(colors=color,direction='in',top=False, right=True, left=False, bottom=False)
    
    axL.spines['right'].set_visible(False)
    axL.tick_params(right=False)
    

    #---------------------------------- Top X-axis (comment out if not needed) -----------------------------------------------
    # axTop = axL.secondary_xaxis('top', (lambda x: x/0.03, lambda x: x*0.03))
    # axTop.set_xlabel("Wavenumber, $\\tilde{\\nu}$ $( \\mathrm{cm^{-1}} )$")
    
    # # Graphic adjustments for top X-axis (don't edit if you don't know what it does)
    # axL.spines['top'].set_visible(False)
    # axTop.tick_params(direction='in')
    # axL.tick_params(top=False)  

    #---------------------------------- Letter annotation ------------------------------------------------------
    letter_annotation(axL, 'E')

drawfigureE(plt.figure(figsize=(4,3)))
```

```python
def drawfigureF(fig):
    # ========================== Loading data from CSV files ===========================================
    T_AEA, TC_AEA = np.loadtxt('./FigureF/AEA.csv', delimiter=',', skiprows=1, unpack=True)
    T_MA, TC_MA = np.loadtxt('./FigureF/MA.csv', delimiter=',', skiprows=1, unpack=True)
    T_WS, TC_WS = np.loadtxt('./FigureF/WS.csv', delimiter=',', skiprows=1, unpack=True)
    T_CS, TC_CS = np.loadtxt('./FigureF/Cs.csv', delimiter=',', skiprows=1, unpack=True)

    ax = fig.subplots(1,1)
    ax.plot(T_AEA, TC_AEA, 'o', label='AEA', markerfacecolor='none',color='C0')
    ax.plot(T_MA, TC_MA, '^', label='MA', markerfacecolor='none',color='C1')
    ax.plot(T_WS, TC_WS, '*', label='WS', markerfacecolor='none',color='C2')
    ax.plot(T_CS, TC_CS, 'p', label='CS', markerfacecolor='none',color='C3')

    ax.set_yscale('log')
    # ax.set_xscale('log')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.tick_params(direction='in', top=True, right=True,which='both')
    # ax.legend(frameon=False, loc='best')
    # ax.text(15, 6.0, 'Signal 1', ha='center', va='center')
    # ax.text(15, 1.0, 'Signal 2', ha='center', va='center')
    ax.annotate('AEA', xy=(T_AEA[10], TC_AEA[10]*1.15), xytext=(T_AEA[0]+30, TC_AEA[0]*1.8), arrowprops=dict(arrowstyle='->', color='C0'), color='C0')
    letter_annotation(ax, 'F')

drawfigureF(plt.figure(figsize=(2,4)))
```

```python
def drawfigureA(fig):
    ax = fig.subplots(1,1)
    img = mpimg.imread('./FigureA/Schematic.png')
    ax.imshow(img)
    ax.axis('off')  # Hide axes
    letter_annotation(ax, 'A')
drawfigureA(plt.figure(figsize=(2,3.0)))
```

```python
def drawfigureD(fig):
    # ========================== Loading data from CSV files ===========================================
    x1 = np.loadtxt('./FigureD/histogram_data.csv', delimiter=',', skiprows=1, unpack=True)
    def normal_distribution(x, mean, std):
        return (1/(std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)
    
    ax1,ax2,ax3 = fig.subplots(1,3)
    fig.subplots_adjust(wspace=0.6)

    nbins = 7
    yshift = 0

    ax1.hist(x1, bins=nbins, color='C0', alpha=0.7, edgecolor='black')

    y_fit =yshift + x1.shape[0] * (x1.max() - x1.min()) / nbins * normal_distribution(np.linspace(x1.min(), x1.max(), 100), np.mean(x1), np.std(x1))
    ax1.plot(np.linspace(x1.min(), x1.max(), 100), y_fit, color='C3')

    ax1.set_xlabel('X-axis')
    ax1.set_ylabel('Y-axis')
    ax1.tick_params(direction='in', top=True, right=True)

    nbins = 7
    yshift = 0

    ax2.hist(x1, bins=nbins, color='C0', alpha=0.7, edgecolor='black')

    y_fit =yshift + x1.shape[0] * (x1.max() - x1.min()) / nbins * normal_distribution(np.linspace(x1.min(), x1.max(), 100), np.mean(x1), np.std(x1))
    ax2.plot(np.linspace(x1.min(), x1.max(), 100), y_fit, color='C3')

    ax2.set_xlabel('X-axis')
    ax2.set_ylabel('Y-axis')
    ax2.tick_params(direction='in', top=True, right=True)

    nbins = 7
    yshift = 0

    ax3.hist(x1, bins=nbins, color='C0', alpha=0.7, edgecolor='black')

    y_fit =yshift + x1.shape[0] * (x1.max() - x1.min()) / nbins * normal_distribution(np.linspace(x1.min(), x1.max(), 100), np.mean(x1), np.std(x1))
    ax3.plot(np.linspace(x1.min(), x1.max(), 100), y_fit, color='C3')

    ax3.set_xlabel('X-axis')
    ax3.set_ylabel('Y-axis')
    ax3.tick_params(direction='in', top=True, right=True)



    # ax.legend(frameon=False, loc='best')
    letter_annotation(ax1, 'D')

drawfigureD(plt.figure(figsize=(4,3)))
```

```python
def drawfigureC(fig):
    #------------------------- Loading data from CSV files ===========================================
    x, y, y_err = np.loadtxt('./FigureC/bar_extracted.csv', delimiter=',', skiprows=1, unpack=True, dtype='object')
    y = y.astype(float)
    y_err = y_err.astype(float)
    
    ax = fig.subplots(1,1)
    ax.bar(x, y, yerr=y_err,color=['C0','C1', 'C2', 'C3'], alpha=0.8, edgecolor='black', width=0.6)
    ax.set_xlabel('')
    ax.set_ylabel('$\\Lambda$ (Wm$^{-1}$K$^{-1}$)')
    ax.tick_params(direction='in', top=True, right=True)
    # ax.text(0.5, 1.05, 'Bar Plot', ha='center', va='center')
    letter_annotation(ax, 'C')
drawfigureC(plt.figure(figsize=(4,3)))
```

```python
figpanel = plt.figure(figsize=(7.2,3.0),constrained_layout=True)

gs = GridSpec(nrows=2, ncols=4, figure=figpanel,width_ratios=[1,1,1,0.7])


subfigA = figpanel.add_subfigure(gs[0, 0])
subfigB = figpanel.add_subfigure(gs[0, 1])
subfigC = figpanel.add_subfigure(gs[0, 2])
subfigD = figpanel.add_subfigure(gs[1, 0:2])
subfigE = figpanel.add_subfigure(gs[1, 2])
subfigF = figpanel.add_subfigure(gs[:, 3])

drawfigureA(subfigA)
drawfigureB(subfigB)
drawfigureC(subfigC)
drawfigureD(subfigD)
drawfigureE(subfigE)
drawfigureF(subfigF)

# Remove the comment (#) to save the figure
# plt.savefig('./fig_panel_final.png', dpi=300,bbox_inches='tight')
# plt.savefig('./fig_panel_final.pdf',format='pdf',bbox_inches='tight')
plt.show()
```
